### 1-)Install required libraries

```python
import pandas as pd
import numpy as np
from geopy.geocoders import Nominatim
import folium
import matplotlib.cm as cm
import matplotlib.colors as colors
```

### 2-)Upload the dataset where we saved it as csv file in part 2 which is called Capstone_part2.csv

```python
df=pd.read_csv('Capstone_part2.csv')
df.head()
```

### 3-)prepare the dataframe in pursuant of instructions in the 3rd part

```python
#Work with only "Borough" which contain the word Toronto 
df1=df[df['Borough'].str.contains('Toronto')]

df1.head()
```

```python
df2=df1.reset_index(drop=True)
df2.head()
```

```python
df2.shape
```

```python
# get  the different unique value of Borough
df2['Borough'].value_counts()
```

```python
#Create a new column as Label
#creat  the date from 'Borough' as integer

df2['Label']=df2['Borough'].replace(to_replace=['Downtown Toronto','Central Toronto','West Toronto','East Toronto',"East York/East Toronto","East Toronto Business","Downtown Toronto Stn A"],value=[1,2,3,4,5,6,7],inplace=False)
df2.head()
```

### 4-)Visualize the data

```python
# the coordinates of Toronto
address = 'Toronto'
geolocator = Nominatim(user_agent="toronto_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print(f'The geograpical coordinate of Toronto are {latitude}, {longitude}.')
```

```python
#for set the cluster number as label number
kclusters=len(df2.Label.unique())

# create map
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, cluster in zip(df2['Latitude'], df2['Longitude'], df2['Label']):
    label = folium.Popup(str(df2['Borough']) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_toronto)
```
